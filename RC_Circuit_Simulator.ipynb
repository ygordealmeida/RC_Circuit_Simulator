```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy import signal
```

```python
def EDO (Vc,t,R,C,Vp):
    
    dvcdt = (Vp-Vc)/(R*C)
    return(dvcdt)
```

```python
#Circuit parameters

Vc0 = 0                  #Inicial Condicion [V]
Vp=5                     #Amplitude    [V]
R=10000                  #Resistance   [Ohms]       
C=100e-6                 #Capacitance  [Faradays]   

t = np.arange(0,7*R*C,0.0001) #Execution time [s]

Vc = odeint(EDO,Vc0,t,args=(R,C,Vp))


#Plotagem da simulao

plt.figure()
plt.plot(t,Vc, label='Capacitor Voltage [V]')
plt.grid()
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.legend()
plt.show()
```
